```python
# Importar los modulos

import pandas as pd
import io
import requests
import seaborn as sns
import timeit
import matplotlib.pyplot as plt
```

IMportamos el dataset

```python
#Cargamos el dataset, lo bajamos de github
# Cargar el dataset
url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
s=requests.get(url).content
titanic=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
titanic
```

```python
#Informmacion del data ser, para ver como vienen los datos
titanic.info()
```

```python
#Renombrar columnas
titanic.rename(columns={'class':'clase'},inplace=True)
titanic.rename(columns={'fare':'tarifa'},inplace=True)

# Mostrar valores unicos (clases)
titanic.clase.unique()

# Primeras 5 filas
titanic.head()
```

```python
# Añadir columna is_old, is_baby

def is_old_func(row):

  return row['age'] > 60


titanic['is_old'] = titanic.apply(is_old_func, axis='columns')


titanic.eval( ' is_baby = age < 15' , inplace = True)
```

```python
# Columna class_num

def class_num_func(row):

  Clase={'Third':3,'First':1,'Second':2}

  return Clase[row.clase]


titanic['class_num'] = titanic.apply(class_num_func, axis='columns')
```

```python
titanic.head()
```

```python
# Query

titanic[(titanic.sex == 'female') & (titanic['clase'].isin(['First','Third'])) & (titanic.age > 45) & (titanic.survived == 0)]
```

```python
#Distribucion variable clase
#cuantas personas hay distribuidas por clase
sns.countplot(x="pclass", data=titanic)
```

```python
#distribucion de pasajeros por edad
#dropna para pasajeros que no tienen registrada la edad, kde pinta linea suavizada
# Distribucion basado en histogramas por la edad
sns.histplot(titanic.age.dropna(), kde=True)
plt.show()
```

```python
# Boxplot edad vs sexo
#hue nos divide la caja en tres según la clase, kind tipo caja
with sns.axes_style(style='ticks'):
  ax = sns.catplot(data=titanic, x="sex", y="age", hue="clase", kind="box")
```

```python
#distribuciones cruzadas
sns.jointplot(x="tarifa", y="age", data=titanic)
```

Interpretación: los dos puntos que están a 500 están en un super camarote o las han timado. Las personas que han paagado más rondan los 20-30. Las personas mayores no pagan una cifra alta.

```python
#matrices de gráficos 
#funcion facetGrid
g = sns.FacetGrid(titanic, row='survived',col='clase')
g.map(sns.histplot, "age", kde=True)
plt.show()
```

```python
#diagrama de dispersión
sns.jointplot(data=titanic, x='age', y='tarifa', kind='reg', color='b')
plt.show()
```

```python
#diagrama de correlaciones
tc = titanic.corr()
sns.heatmap(tc,cmap='coolwarm')
plt.title('titanic.corr()')
```
